Deberas crear tu propia funci칩n para calcular el error de reconstrucci칩n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
import numpy as np

def sse_score(X, X_hat):
    
    SSE = np.sum((X - X_hat)**2)
    
    return SSE
```

Una vez que ya tenemos la funci칩n `sse` hecha, podemos pasar a construir la funci칩n que ejecutar치 `SVD`.

```python
def svd(X):
    
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    S = np.diag(s)
   
    return U, S, Vt
```

Como hemos visto en clase, las matrices obtenidas a partir de `SVD` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci칩n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
def reconstruction(u, s, vt):
    
    A_hat = np.dot(u, np.dot(s, vt))
    
    return A_hat
```

Una vez que hemos programado todas las funciones necesarias para realizar `SVM` y medir el error de reconstrucci칩n, podemos proceder a realizar la compresi칩n de la imagen. Esta [p치gina web](http://timbaumann.info/svd-image-compression-demo/) te ayudar치 a repasar y a entender como calcular la compresi칩n.

Debes usar la siguiente imagen: 

```python
import matplotlib.pyplot as plt
from scipy import misc
%matplotlib inline

# Load image
A = misc.face(gray=True)

plt.imshow(A, cmap=plt.cm.gray)
```

```python
def image_compression(A, n_components):

    U, S, Vt = svd(A)
    
    reduced_U = U[:, :n_components]        
    reduced_S = S[:n_components, :n_components]
    reduced_Vt = Vt[:n_components, :]
    
    A_hat = reconstruction(reduced_U, reduced_S, reduced_Vt)
    sse = sse_score(A, A_hat)
    comp_ratio = (A.shape[1]*n_components + n_components + n_components*A.shape[0]) / (A.shape[1] * A.shape[0])
    
    return A_hat, sse, comp_ratio
```

```python
A_hat, sse, comp_ratio = image_compression(A, n_components=50)
```

### Regresi칩n Lineal

Ya que conocemos la ecuaci칩n para $w_{LS}$ tenemos todo lo necesario para resolver la regresi칩n lineal. Vamos all치!<br><br>

<center>$w_{LS} = (X^T X)^{-1}X^T y,$</center>


```python
def inverse_of_matrix(mat):
    
    mat_inv = np.linalg.inv(mat)
    
    return mat_inv
```

#### Subset del dataframe por valores

El siguiente paso es construir una funci칩n que recibe un `data_frame`, el nombre de una columna, un valor m칤nimo y un valor m치ximo `cutoffs`. Nos devuelve un dataframe excluyendo las filas donde el valor de la columna indica est치 fuera de los valores m칤nimos y m치ximos que le hemos indicado.

```python
def column_cutoff(data_frame, cutoffs):
    
    #df_sub_cutoff = data_frame[(data_frame[cutoffs[0][0]] > cutoffs[0][1]) & (data_frame[cutoffs[1][0]] < cutoffs[1][2])]
        
    df_sub_cutoff = data_frame[(data_frame[cutoffs[0][0]].between(cutoffs[0][1],cutoffs[0][2],inclusive="neither")) & (data_frame[cutoffs[1][0]].between(cutoffs[1][1],cutoffs[1][2],inclusive="neither"))]
    
    return df_sub_cutoff
```

#### M칤nimos Cuadrados / Least Squares

Ahora, implementar치s la ecuaci칩n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
def least_squares_weights(input_x, target_y):
    
    if input_x.shape[0] < input_x.shape[1]:
        input_x = np.transpose(input_x)
    if target_y.shape[0] < target_y.shape[1]:
        target_y = np.transpose(target_y)
     
    input_x = np.c_[np.ones(input_x.shape[0]),input_x]
    xt = np.transpose(input_x)
    inv_x = inverse_of_matrix(np.dot(xt, input_x))
    
    weights = np.dot(np.dot(inv_x, xt), target_y)
  
    return weights
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem치ticos detr치s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi칩n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m치ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi칩n lineal, se basa en resolver esta ecuaci칩n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('./data/train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Contruir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x] 

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
def gradient_descent(x, y, w, iterations, eta):

    past_loss = []
    past_w = [w]
    n = y.size
    for i in range(iterations):
        prediction = np.dot(x, w)
        error = prediction - y
        
        # TODO: Define Loss function
        #=============
        
        loss = (1/n) * np.sum((y - np.dot(x, past_w[i]))**2)
        
        #=============
        past_loss.append(loss)
        
        GradRss = np.dot(x.T, error)
        
        # TODO: Define updating rule
        #=============
        
        w = past_w[i] - eta*GradRss
        
        #=============
        past_w.append(w)
        
    return past_w, past_loss
```

Una vez construida nuestra funci칩n para el Gradient Descent podemos usarla para encontrar los valores optimos de $w$. **Prueba a modificar el learning rate para ver la convergencia del Gradient Descent.**

```python
import numpy as np

# Learning rate
eta = 0.001
#eta = 0.00001

# N칰mero de iteraciones
iterations = 2000 #No. of iterations

# Seed para inicializar w
np.random.seed(123)
w0 = np.random.rand(2)

weights, loss = gradient_descent(x, y, w0, iterations, eta)

print(weights[-1])
```

## (Opcional) - Calculando similitud entre p치ginas web

### Repaso

Como recordar치s, podemos medir la similitud entre vectores usando la siguiente ecuaci칩n:<br>

<center>$\overrightarrow{u} \cdot \overrightarrow{v} = |\overrightarrow{u}||\overrightarrow{v}| \cos \theta $</center>

Que podemos reescribir de la siguiente forma:<br>

<center>$\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$</center>

La **similitud** va a venir dada por el 치ngulo $\theta$, que nos indicar치 lo siguiente:

```python
import re
import lxml
from bs4 import BeautifulSoup
import urllib
import urllib.request

def parse_from_url(url):

    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, "html.parser" )
    for script in soup(["script", "style"]):
        script.extract()
        
    text = soup.get_text()
    
    # Eliminar saltos de linea
    text = re.sub('\s+', ' ', text)
    return text
```

```python
def get_content(url_ls):
    
    url2content = {}
    
    for url in url_ls:
        value = parse_from_url(url)
        url2content.update({url: value})     
    
    return url2content
```

### Preprocesado

Como es l칩gico, no podemos resolver esta ecuaci칩n $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$ usando texto sin m치s, debemos convertir cada p치gina web a un vector.

```python
def tokenizer(text):
    
    tokens_txt = text.split(" ")
    set_txt = set()

    for item in tokens_txt:
        if item not in set_txt:
            set_txt.add(item)        

    return tokens_txt, set_txt
```

El siguiente paso es crear un conjunto con las palabras de ambas p치ginas web (uni칩n), por ejemplo:

- Los perros son maravillosos...
- Los maravillosos a침os 80...

Por tanto, el conjunto para estas dos frases ser칤a `{"los", "perros", "son", "maravillosos", "a침os", "80"}`. Debemos realizar esto para todas las combinaciones posibles, es decir:

- web_1
- web_2
- web_3

En este caso, las combinaciones ser칤an (no importa el orden) `[web_1, web_2]`, `[web_1, web_3]`, `[web_2, web_3]`

```python
import itertools
```

```python
import itertools
from itertools import combinations
    
def combinationsURL(url_ls):
    
    comb_ls_url = list(combinations(url_ls, 2))
    
    return comb_ls_url
```

```python
def union(comb_ls):
    
    new_list_combi = []
    
    for c in comb_ls:
        join_set = c[0].union(c[1])
        new_list_combi.append(join_set)

    return new_list_combi
```

Una vez que tenemos una lista de conjuntos por cada par de p치ginas web, podemos convertir el texto de la p치gina web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)        

    return web1_array, web2_array
```

### Cosine Similarity

Por 칰ltimo, ya podemos implementar la ecuaci칩n: $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$

```python
import numpy as np

def cosine_similarity(u, v):
    
    u = np.array(u)
    v = np.array(v)
    
    similiraty = (u @ v) / (np.linalg.norm(u)*np.linalg.norm(v))

    return similiraty
```

```python
def websites_sim(url_ls):

    url2content = get_content(url_ls)
    
    # Creamos un diccionario donde cada url tendr치 su contenido tokenizado y su conjunto
    url_dict = {}
    for url, content in url2content.items():
        toks, sets = tokenizer(content)
        url_dict[url] = {'tokens': toks,
                        'unique_tokens': sets}
    
    # Calculamos todas las combinaciones posibles de las direcciones de las p치ginas web
    comb_ls = combinationsURL(url_ls)

    # Usando comb_ls y la funci칩n `set2vector()` convertimos cada p치gina web a vectores
    print("Similaridad: ")
    for el in comb_ls:
        # Obtenemos los tokens y el conjunto para cada p치gina web
        token_1 = url_dict[el[0]]['tokens']
        token_2 = url_dict[el[1]]['tokens']
        set_1 = url_dict[el[0]]['unique_tokens']
        set_2 = url_dict[el[1]]['unique_tokens']
        array_web1, array_web2 = set2vector(token_1, token_2, set_1, set_2)
        similarity = cosine_similarity(array_web1, array_web2)
        
        print("{} vs {} - {}".format(el[0], el[1], round(similarity, 3)))
                      
```
